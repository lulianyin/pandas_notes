## 【任务一】企业收入的多样性
<img src="../source/_static/期中-任务1.png" width="100%">

```python
df1 = pd.read_csv('../data/测试/Company_data.csv')
df1.head()
```

```python
df2 = pd.read_csv('../data/测试/Company.csv')
df2.head()
```

```python
df1.dtypes
```

```python
# df1 聚类，按照证券代码、年份、收入类型
df1['日期'] = df1['日期'].str[:4]
group_df1 = df1.groupby(['证券代码','日期','收入类型']).sum()
group_df1.reset_index(inplace=True)
group_df1.head()
```

```python
# 求和，计算比例
group_df1['总和'] = group_df1.groupby(['证券代码','日期'])['收入额'].transform('sum')
group_df1['比例'] = group_df1['收入额']/group_df1['总和']
group_df1.head(5)
```

```python
# log
group_df1['信息'] = group_df1['比例'] * np.log(group_df1['比例'])
group_df1['证券代码'] = group_df1['证券代码'].apply(lambda x:'#' + ('000000'+str(x))[-6:])
group_df1.head(5)
```

```python
# 聚类，更新到df2
group2_df1 = group_df1.groupby(['证券代码','日期'])['信息'].sum().reset_index()
group2_df1['日期'] = group2_df1['日期'].astype('int')
```

```python
df2.merge(group2_df1,on=['证券代码','日期'],how='left')
```

## 【任务三】美国大选投票情况

【题目描述】两张数据表中分别给出了美国各县（county）的人口数以及大选的投票情况，请解决以下问题：

1. 有多少县满足总投票数超过县人口数的一半
2. 把州（state）作为行索引，把投票候选人作为列名，列名的顺序按照候选人在全美的总票数由高到低排序，行列对应的元素为该候选人在该州获得的总票数
<img src="../source/_static/期中-任务3.png" width="80%">
3. 每一个州下设若干县，定义拜登在该县的得票率减去川普在该县的得票率为该县的BT指标，若某个州所有县BT指标的中位数大于0，则称该州为Biden State，请找出所有的Biden State

```python
df4 = pd.read_csv('../data/测试/president_county_candidate.csv')
df4.head()
```

```python
df5 = pd.read_csv('../data/测试/county_population.csv')
df5.head()
```

### 问题1

```python
# df4 聚类
df4_sum = pd.DataFrame(df4.groupby(['county'])['total_votes'].sum())
df4_sum.head(3)
```

```python
df5['US County'] = df5['US County'].str[1:]
df5['County'] = df5['US County'].str.split(',',expand=True)[0]
df5.head()
```

```python
ans1 = df5.merge(df4_sum, left_on='County', right_on='county', how='left')
ans1.head(3)
```

```python
ans1[ans1['total_votes']/ans1['Population']>0.5]
```

### 问题2

```python
# 通过县透视
pivot_df4 = df4.pivot(index = ['state', 'county'],
                       columns = ['candidate'],
                       values = ['total_votes'])
pivot_df4.reset_index(inplace=True)
pivot_df4 = pivot_df4.droplevel(0,axis=1) 
pivot_df4.columns = ['state','county']+list(pivot_df4.columns[2:])
pivot_df4.head()
```

```python
ans2 = pivot_df4.groupby('state').sum()
```

```python
# 排序得到结果
ans2 = ans2[list(ans2.sum().sort_values(ascending = False).index)]
ans2.head()
```

### 问题3

```python
# 问题2的基础上，求解县的BT指标
# 每一个州下设若干县，定义拜登在该县的得票率减去川普在该县的得票率为该县的BT指标，若某个州所有县BT指标的中位数大于0，则称该州为Biden State，请找出所有的Biden State
pivot_df4.head()
ans3 = pivot_df4[['Joe Biden','Donald Trump','state','county']].merge(df5, left_on='county', right_on='County', how='left')
ans3['BT'] = ans3['Joe Biden']/ans3['Population']-ans3['Donald Trump']/ans3['Population']
ans3.head(3)
```

```python
# 查找中位数, 注意有县人数未知，这里认为BT=0
ans3['BT'] = ans3['BT'].fillna(0)
ans3 = ans3.groupby(['state']).median()
```

```python
ans3['Biden State'] = False
ans3[ans3['BT'] >0] = True
```

```python
ans3[ans3['BT'] > 0].index   # ans
```
