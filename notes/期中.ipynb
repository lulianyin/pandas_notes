```python
df4 = pd.read_csv('../data/æµ‹è¯•/president_county_candidate.csv')
df4.head()
```

```python
df5 = pd.read_csv('../data/æµ‹è¯•/county_population.csv')
df5.head()
```

### é—®é¢˜1

```python
# df4 èšç±»
df4_sum = pd.DataFrame(df4.groupby(['county'])['total_votes'].sum())
df4_sum.head(3)
```

```python
df5['US County'] = df5['US County'].str[1:]
df5['County'] = df5['US County'].str.split(',',expand=True)[0]
df5.head()
```

```python
ans1 = df5.merge(df4_sum, left_on='County', right_on='county', how='left')
ans1.head(3)
```

```python
ans1[ans1['total_votes']/ans1['Population']>0.5]
```

### é—®é¢˜2

```python
# é€šè¿‡å¿é€è§†
pivot_df4 = df4.pivot(index = ['state', 'county'],
                       columns = ['candidate'],
                       values = ['total_votes'])
pivot_df4.reset_index(inplace=True)
pivot_df4 = pivot_df4.droplevel(0,axis=1) 
pivot_df4.columns = ['state','county']+list(pivot_df4.columns[2:])
pivot_df4.head()
```

```python
ans2 = pivot_df4.groupby('state').sum()
```

```python
# æ’åºå¾—åˆ°ç»“æœ
ans2 = ans2[list(ans2.sum().sort_values(ascending = False).index)]
ans2.head()
```

### é—®é¢˜3

```python
# é—®é¢˜2çš„åŸºç¡€ä¸Šï¼Œæ±‚è§£å¿çš„BTæŒ‡æ ‡
# æ¯ä¸€ä¸ªå·ä¸‹è®¾è‹¥å¹²å¿ï¼Œå®šä¹‰æ‹œç™»åœ¨è¯¥å¿çš„å¾—ç¥¨ç‡å‡å»å·æ™®åœ¨è¯¥å¿çš„å¾—ç¥¨ç‡ä¸ºè¯¥å¿çš„BTæŒ‡æ ‡ï¼Œè‹¥æŸä¸ªå·æ‰€æœ‰å¿BTæŒ‡æ ‡çš„ä¸­ä½æ•°å¤§äº0ï¼Œåˆ™ç§°è¯¥å·ä¸ºBiden Stateï¼Œè¯·æ‰¾å‡ºæ‰€æœ‰çš„Biden State
pivot_df4.head()
ans3 = pivot_df4[['Joe Biden','Donald Trump','state','county']].merge(df5, left_on='county', right_on='County', how='left')
ans3['BT'] = ans3['Joe Biden']/ans3['Population']-ans3['Donald Trump']/ans3['Population']
ans3.head(3)
```

```python
# æŸ¥æ‰¾ä¸­ä½æ•°, æ³¨æ„æœ‰å¿äººæ•°æœªçŸ¥ï¼Œè¿™é‡Œè®¤ä¸ºBT=0
ans3['BT'] = ans3['BT'].fillna(0)
ans3 = ans3.groupby(['state']).median()
```

```python
ans3['Biden State'] = False
ans3[ans3['BT'] >0] = True
```
